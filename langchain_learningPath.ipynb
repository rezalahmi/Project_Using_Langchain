```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size=200,
    chunk_overlap=0
)
```

```python
chunks = splitter.split_text(data[0].page_content)
len(chunks)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
r_splitter = RecursiveCharacterTextSplitter(
    separators = ["\n\n", "\n", ".", ",", " "],  # List of separators based on requirement (defaults to ["\n\n", "\n", " "])
    chunk_size = 200,  # size of each chunk created
    chunk_overlap  = 0,  # size of  overlap between chunks in order to maintain the context
    length_function = len  # Function to calculate size, currently we are using "len" which denotes length of string however you can pass any token counter)
)
```

```python
chunks = r_splitter.split_text(data[0].page_content)
```

```python
len(chunks)
```

```python
import time
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By

# Launch Chrome browser in headless mode
options = webdriver.ChromeOptions()
# options.add_argument("headless")
browser = webdriver.Chrome(options=options)

# Load web page
browser.get(url="https://fa.wikipedia.org/wiki/%D8%AF%D8%A7%D8%B3%D8%AA%D8%A7%D9%86_%DA%A9%D9%88%D8%AA%D8%A7%D9%87")
# Save the page HTML to a file
with open("index.html", "w", encoding="utf-8") as file:
    file.write(browser.page_source)

# Close the browser once finish
browser.close()
```

```python
from langchain.document_loaders import UnstructuredHTMLLoader

# Load the HTML file using UnstructuredHTMLLoader
print('Load the HTML file')
loader = UnstructuredHTMLLoader("index.html")

# Process the loaded data
# The specific method to process or extract data depends on the UnstructuredHTMLLoader's functionality
# For example, you might be able to extract text or other elements depending on the loader's capabilities
print('Process the loaded data')
processed_data = loader.load()  # or any other method provided by the loader

# Do something with the processed data
# For example, print it, analyze it, store it, etc.
print(processed_data)
```
